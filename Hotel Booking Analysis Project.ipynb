```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def replacer(df):
    cat = list(df.columns[df.dtypes == "object"])

    for i in df.columns:
        if i in cat:
            mode = df[i].mode()[0]
            df[i] = df[i].fillna(mode)
        else:
            mean = df[i].mean()
            df[i] = df[i].fillna(mean)
```

```python
replacer(df)
```

```python
df.isnull().sum()
```

```python
cat = []
con = []

for i in df.columns:
    if df[i].dtypes == "object":
        cat.append(i)
    else:
        con.append(i)
```

```python
df["reservation_status_date"] = pd.to_datetime(df["reservation_status_date"])
```

```python
cancellation_rate = df["is_canceled"].mean()*100
print(f"Cancellation_Rate:{cancellation_rate:.2f}%")
```

```python
# Bookings Per Month
monthly_bookings = df.groupby("arrival_date_month")["is_canceled"].count().sort_values()
plt.figure(figsize = (15, 5))
sns.barplot(x = monthly_bookings.index, y = monthly_bookings.values, palette = "coolwarm", edgecolor = "black")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Bookings Per Month")
plt.xticks(rotation = 45)
plt.show()
```

```python
# Market Segment Analysis
plt.figure(figsize = (15, 5))
sns.countplot(y = df["market_segment"], order = df["market_segment"].value_counts().index, palette = "Set2", edgecolor = "black")
plt.xlabel("Number of Bookings")
plt.ylabel("Market Segments")
plt.title("Bookings by Market Segment")
plt.show()
```

```python
# Distribution Channel Analysis
plt.figure(figsize = (15, 5))
sns.countplot(y = df["distribution_channel"], order = df["distribution_channel"].value_counts().index, palette = "coolwarm", edgecolor = "black")
plt.xlabel("Number of Bookings")
plt.ylabel("Distribution Channel")
plt.title("Bookings by Distribution Channel")
plt.show()
```

```python
# Lead Time vs Cancellation
plt.figure(figsize = (15, 5))
sns.boxplot(x = df["is_canceled"], y = df["lead_time"], palette = "rocket")
plt.xlabel("Is Canceled (1 = Yes, 0 = No)")
plt.ylabel("Lead Time (Days)")
plt.title("Lead Time and Cancellation Relationship")
plt.show()
```

```python
# Average Daily Rate(ADR) Analysis
plt.figure(figsize = (15, 5))
sns.histplot(df["adr"], bins = 50, kde = True)
plt.xlabel("ADR (Average Daily Rate)")
plt.ylabel("Frequency")
plt.title("ADR & Frequency")
plt.show()
```

```python
# Special Requests and Cancellations
plt.figure(figsize = (15, 5))
sns.boxplot(x = df["is_canceled"], y = df["total_of_special_requests"], palette = "coolwarm")
plt.xlabel("Is Canceled (1 = Yes, 0 = No)")
plt.ylabel("Total Special Requests")
plt.title("Total Requests and Cancellation Relationship")
plt.show()
```

```python
# A) Booking Trends Over Time
# Convert Month Names to Numerical Format

df["arrival_date_month"] = pd.Categorical(df["arrival_date_month"],
                                                categories = ["January", "February", "March", "April", "May", "June",
                                                              "July", "August", "September", "October", "November", "December"],
                                                ordered = True)

# Plot Bookings Per Month
plt.figure(figsize = (15, 5))
sns.countplot(x = df["arrival_date_month"], order = df["arrival_date_month"].cat.categories, palette = "Set2", edgecolor = "black")
plt.xlabel("Month")
plt.ylabel("Number of Bookings")
plt.title("Monthly Booking Trends")
plt.show()
```

```python
# B) Market Segment Analysis
plt.figure(figsize = (15,5))
sns.countplot(y = df["market_segment"], order = df["market_segment"].value_counts().index, palette = "viridis", edgecolor = "black")
plt.xlabel("Number of Bookings")
plt.ylabel("Market Segments")
plt.title("Bookings by Market Segment")
plt.show()
```

```python
# C) Booking Source Analysis
plt.figure(figsize = (15, 5))
sns.countplot(y = df["distribution_channel"], order = df["distribution_channel"].value_counts().index, palette = "rocket", edgecolor = "black")
plt.xlabel("Number of Bookings")
plt.ylabel("Distribution channel")
plt.title("Bookings by Distribution Channel")
plt.show()
```

```python
# D) Cancellation Rate Analysis
cancellation_rate = df["is_canceled"].mean()*100
print(f"Cancellation Rate:{cancellation_rate:.2f}%")
```

```python
plt.figure(figsize = (15, 5))
sns.countplot(x = df["is_canceled"], palette = "coolwarm", edgecolor = "black")
plt.xlabel("Is Canceled? (1 = Yes, 0 = No)")
plt.ylabel("Count")
plt.title("Booking Cancellation")
plt.xticks([0, 1], ["Not Canceled", "Canceled"])
plt.show()
```

```python
# E) Lead Time and Cancellations
plt.figure(figsize = (15, 5))
sns.boxplot(x = df["is_canceled"], y = df["lead_time"], palette = "Set2")
plt.xlabel("Is_Canceled (1 = Yes, 0 = No)")
plt.ylabel("Lead Time (Days)")
plt.title("Lead Time and Cancellations")
plt.show()
```

```python
# F) Pricing (ADR - Average Daily Rate) Impact
plt.figure(figsize = (15, 5))
sns.histplot(df["adr"], bins = 50, kde = True, color = "blue")
plt.xlabel("ADR (Average Daily Rate)")
plt.ylabel("Frequency")
plt.title("ADR Distribution")
plt.show()
```

```python
# G) Special Requests and Cancellations
plt.figure(figsize = (15, 5))
sns.boxplot(x = df["is_canceled"], y = df["total_of_special_requests"], palette = "coolwarm")
plt.xlabel("Is Canceled (1 = Yes, 0 = No)")
plt.ylabel("Total Special Requests")
plt.title("Special Requests and Cancellations")
plt.show()
```

```python
top_months = df["arrival_date_month"].value_counts().nlargest(5).index
filtered_df = df[df["arrival_date_month"].isin(top_months)].copy()
top_companies = filtered_df["company"].value_counts().nlargest(5).index
filtered_df = filtered_df[filtered_df["company"].isin(top_companies)]
ctab = pd.crosstab(filtered_df["arrival_date_month"], filtered_df["company"])

plt.figure(figsize = (12, 5))
sns.heatmap(ctab, annot = True, fmt = "d", cmap = "coolwarm")
plt.title("Top 5 Arrival Month vs Top 5 Companies")
plt.xlabel("Company")
plt.ylabel("Arrival Date Month")
plt.show()
```
